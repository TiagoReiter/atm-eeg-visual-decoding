```python
import wandb
from dateutil.parser import parse as parse_date
import os

# Make sure your API key & mode are set
os.environ["WANDB_API_KEY"] = "fc2d1bc24195ed0dc256cf0d1a94a44630eff1e7"
os.environ["WANDB_MODE"]    = "online"

api  = wandb.Api()
runs = api.runs("t-reiter-technical-university-of-munich/Forschungspraxis")

# 1) Filter client‐side: only in‐subject sub-08, names starting with "dm"
phase2 = []
for r in runs:
    cfg = r.config
    subjects = cfg.get("subjects", [])
    if isinstance(subjects, str):
        try: subjects = eval(subjects)
        except: subjects = [subjects]
    if not (cfg.get("insubject") and "sub-08" in subjects):
        continue
    if r.name.startswith("dm"):
        phase2.append(r)

if not phase2:
    print("No Phase 2 runs found for sub-08!")
    exit()

# 2) Find the best
best_run, best_acc = None, -1.0
for r in phase2:
    acc = r.summary.get("Test Accuracy") or r.summary.get("test_accuracy") or 0.0
    if acc > best_acc:
        best_acc = acc
        best_run = r

# 3) Report the best
cfg = best_run.config
print("⭐️ Best Phase 2 run:")
print(f"   name       = {best_run.name}")
print(f"   created_at = {best_run.created_at}")
print(f"   Test Acc   = {best_acc:.4f}")
print(f"   d_model    = {cfg.get('d_model')}")
print(f"   n_heads    = {cfg.get('n_heads')}")
print(f"   e_layers   = {cfg.get('e_layers')}")
print("")

# 4) Print all Phase 2 runs, sorted by Test Accuracy descending
print("📋 All Phase 2 runs (sorted by Test Acc):")
sorted_runs = sorted(
    phase2,
    key=lambda r: r.summary.get("Test Accuracy") or r.summary.get("test_accuracy") or 0.0,
    reverse=True
)
for r in sorted_runs:
    acc      = r.summary.get("Test Accuracy") or r.summary.get("test_accuracy") or 0.0
    cfg      = r.config
    dm       = cfg.get("d_model")
    nh       = cfg.get("n_heads")
    el       = cfg.get("e_layers")
    created  = r.created_at
    print(f"→ {r.name:15s} | Test Acc={acc:.4f} | d_model={dm} | n_heads={nh} | e_layers={el} | created={created}")
```

```python
import wandb
import os

# Setup
os.environ["WANDB_API_KEY"] = "fc2d1bc24195ed0dc256cf0d1a94a44630eff1e7"
os.environ["WANDB_MODE"]    = "online"

api  = wandb.Api()
runs = api.runs("t-reiter-technical-university-of-munich/Forschungspraxis")

# 1) Filter client-side: only runs from batch_size / epochs sweep (bs*_ep*)
batch_epoch_runs = []
for r in runs:
    if r.name.startswith("bs") and "_ep" in r.name and r.config.get("encoder_type") == "ATMS":
        batch_epoch_runs.append(r)

if not batch_epoch_runs:
    print("❌ No batch_size/epochs sweep runs found!")
    exit()

# 2) Find the best
best_run, best_acc = None, -1.0
for r in batch_epoch_runs:
    acc = r.summary.get("Test Accuracy") or r.summary.get("test_accuracy") or 0.0
    if acc > best_acc:
        best_acc = acc
        best_run = r

# 3) Report best run
cfg = best_run.config
print("⭐️ Best ATMS Batch/Epoch Run:")
print(f"   name       = {best_run.name}")
print(f"   created_at = {best_run.created_at}")
print(f"   Test Acc   = {best_acc:.4f}")
print(f"   batch_size = {cfg.get('batch_size')}")
print(f"   epochs     = {cfg.get('epochs')}")
print()

# 4) Print all batch/epoch runs sorted by Test Accuracy
print("📋 All Batch/Epoch Runs (sorted by Test Accuracy):")
sorted_runs = sorted(
    batch_epoch_runs,
    key=lambda r: r.summary.get("Test Accuracy") or r.summary.get("test_accuracy") or 0.0,
    reverse=True
)
for r in sorted_runs:
    acc      = r.summary.get("Test Accuracy") or r.summary.get("test_accuracy") or 0.0
    cfg      = r.config
    bs       = cfg.get("batch_size")
    epochs   = cfg.get("epochs")
    created  = r.created_at
    print(f"→ {r.name:12s} | Test Acc={acc:.4f} | batch_size={bs} | epochs={epochs} | created={created}")
```
